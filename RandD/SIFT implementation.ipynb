```python
import numpy as np

```

```python
def gaussian_filter(sigma): 
  size = 2*np.ceil(3*sigma)+1 
  x, y = np.mgrid[-size//2 + 1:size//2 + 1, -size//2 + 1:size//2 + 1] 
  g = np.exp(-((x**2 + y**2)/(2.0*sigma**2))) / (2*np.pi*sigma**2)
  return g/g.sum()
```

```python
gaussian_filter(5)
```
